### Unir Previsões com Dados Tabulares


```python
def unir_datasets(tipo):
    index = []
    prob1 = []
    prob2 = []
    pasta = f'./BBAS3.SA/imagens/{tipo}/vender'
    for i in os.listdir(pasta):
        proba = classifier.predict_image(f'{pasta}/{i}')
        index.append(i.split("_")[0])
        prob1.append(float(proba[0][0]))
        prob2.append(float(proba[0][1]))

    pasta = f'./BBAS3.SA/imagens/{tipo}/comprar'
    for i in os.listdir(pasta):
        proba = classifier.predict_image(f'{pasta}/{i}')
        index.append(i.split("_")[0])
        prob1.append(float(proba[0][0]))
        prob2.append(float(proba[0][1]))

    
    dataset = pd.read_csv(f'PETR4.SA/{tipo}.csv', index_col='Date')
    dataset_images = pd.DataFrame()
    dataset_images['Date'] = index
    dataset_images['Prob_Image_Compra'] = prob1
    dataset_images['Prob_Image_Venda'] = prob2

    dataset_images.set_index('Date')
    dataset_unido = pd.merge(dataset, dataset_images, on='Date', how='inner')
    dataset_unido['Label'] = dataset_unido['Label'].apply(lambda x: x if int(x) == 1 else 0)
    return dataset_unido
```

Um novo **DataFrame com dados de mercado + previsões baseadas em imagens**, que pode ser usado para:
- Análise exploratória.
- Treinamento de um modelo tabular com features visuais e numéricas.
- Geração de estratégias de investimento mais robustas.

> **Nota**: Esse processo representa uma integração típica entre **Visão Computacional** e **Machine Learning Tabular**, muito útil em aplicações como **trading algorítmico**, **detecção de padrões gráficos** e **sistemas de recomendação financeira**.

```python
dataset_treino = unir_datasets('treino')
dataset_teste = unir_datasets('teste')
```

### Preparar Dados Tabulares para o Classificador

#### Features Selecionadas (`X`):
- `Close`: Preço de fechamento do dia.
- `Smoothed_Close`: Versão suavizada do preço de fechamento.
- `Past_1_Days_Close` a `Past_15_Days_Close`: Preços de fechamento dos últimos 15 dias anteriores, servindo como histórico temporal.


```python
X = dataset_treino[['Past_1_Days_Close', 'Past_2_Days_Close', 'Past_3_Days_Close',
       'Past_4_Days_Close', 'Past_5_Days_Close', 'Past_6_Days_Close',
       'Past_7_Days_Close', 'Past_8_Days_Close', 'Past_9_Days_Close',
       'Past_10_Days_Close', 'Past_11_Days_Close', 'Past_12_Days_Close',
       'Past_13_Days_Close', 'Past_14_Days_Close', 'Past_15_Days_Close']]  
y = dataset_treino['Label']
```

```python
X_test = dataset_teste[['Past_1_Days_Close', 'Past_2_Days_Close', 'Past_3_Days_Close',
       'Past_4_Days_Close', 'Past_5_Days_Close', 'Past_6_Days_Close',
       'Past_7_Days_Close', 'Past_8_Days_Close', 'Past_9_Days_Close',
       'Past_10_Days_Close', 'Past_11_Days_Close', 'Past_12_Days_Close',
       'Past_13_Days_Close', 'Past_14_Days_Close', 'Past_15_Days_Close']] 
y_test = dataset_teste['Label']
```

## Análise do Ensemble Simples (Tabular + Imagem)

###  Objetivo

Combinar as previsões do modelo tabular e do modelo baseado em imagem para melhorar a robustez e acurácia da classificação entre as classes **"comprar" (1)** e **"vender" (0)**.

---

###  Estratégia do Ensemble

A abordagem utilizada foi baseada na **soma das probabilidades**:

- Para cada amostra, somamos as probabilidades de **compra** e **venda** vindas de:
  - Modelo tabular (dados estruturados)
  - Modelo baseado em imagem

 **Decisão final**: Se a soma das probabilidades de **compra** for maior que a de **venda**, classificamos como `1` (comprar); caso contrário, como `0` (vender).


```python
y_pred_proba = IModel.model.predict(X)
dataset_previsoes_treino = pd.DataFrame()
dataset_previsoes_treino["y_previsto_preco_prob_venda"] = [float(x[0]) for x in y_pred_proba]
dataset_previsoes_treino["y_previsto_preco_prob_compra"] = [float(x[1]) for x in y_pred_proba]
dataset_previsoes_treino["Prob_Image_Compra"] = dataset_treino['Prob_Image_Compra']
dataset_previsoes_treino["Prob_Image_Venda"] = dataset_treino['Prob_Image_Venda']
dataset_previsoes_treino["label"] = dataset_treino['Label']

dataset_previsoes_treino["Soma_Prob_Compra"] = (
    dataset_previsoes_treino["y_previsto_preco_prob_compra"] + 
    dataset_previsoes_treino["Prob_Image_Compra"]
)

dataset_previsoes_treino["Soma_Prob_Venda"] = (
    dataset_previsoes_treino["y_previsto_preco_prob_venda"] + 
    dataset_previsoes_treino["Prob_Image_Venda"]
)

# Aplica a regra: se soma de compra for maior, prever compra (1); senão, prever venda (0)
dataset_previsoes_treino["y_pred_ensemble_simples"] = dataset_previsoes_treino.apply(
    lambda row: 1 if row["Soma_Prob_Compra"] > row["Soma_Prob_Venda"] else 0,
    axis=1
)
y_true = dataset_previsoes_treino["label"]
y_pred_ensemble = dataset_previsoes_treino["y_pred_ensemble_simples"]

print("Accuracy do Ensemble Simples:", accuracy_score(y_true, y_pred_ensemble))
print("Confusion Matrix:\n", confusion_matrix(y_true, y_pred_ensemble))
print("Classification Report:\n", classification_report(y_true, y_pred_ensemble))
```

```python
y_pred_proba = IModel.model.predict(X_test)
dataset_previsoes_teste = pd.DataFrame()
dataset_previsoes_teste["y_previsto_preco_prob_venda"] = [float(x[0]) for x in y_pred_proba]
dataset_previsoes_teste["y_previsto_preco_prob_compra"] = [float(x[1]) for x in y_pred_proba]
dataset_previsoes_teste["Prob_Image_Compra"] = dataset_teste['Prob_Image_Compra']
dataset_previsoes_teste["Prob_Image_Venda"] = dataset_teste['Prob_Image_Venda']
dataset_previsoes_teste["label"] = dataset_teste['Label']
# Calcula a soma das probabilidades de compra e venda
dataset_previsoes_teste["Soma_Prob_Compra"] = (
    dataset_previsoes_teste["y_previsto_preco_prob_compra"] + 
    dataset_previsoes_teste["Prob_Image_Compra"]
)

dataset_previsoes_teste["Soma_Prob_Venda"] = (
    dataset_previsoes_teste["y_previsto_preco_prob_venda"] + 
    dataset_previsoes_teste["Prob_Image_Venda"]
)

# Aplica a regra: se soma de compra for maior, prever compra (1); senão, prever venda (0)
dataset_previsoes_teste["y_pred_ensemble_simples"] = dataset_previsoes_teste.apply(
    lambda row: 1 if row["Soma_Prob_Compra"] > row["Soma_Prob_Venda"] else 0,
    axis=1
)
y_true = dataset_previsoes_teste["label"]
y_pred_ensemble = dataset_previsoes_teste["y_pred_ensemble_simples"]

print("Accuracy do Ensemble Simples:", accuracy_score(y_true, y_pred_ensemble))
print("Confusion Matrix:\n", confusion_matrix(y_true, y_pred_ensemble))
print("Classification Report:\n", classification_report(y_true, y_pred_ensemble))
```
